# Perform applied NLP 
- Using the ebook - The life story of a squirrel

## Obtain Data 

```python
import requests 

url = 'https://raw.githubusercontent.com/Mund99/DSUB_5.0/main/Dataset/The%20life%20story%20of%20a%20squirrel.txt'

response = requests.get(url)
text = response.text 
```

```python
print(text)
```

## Scrub Data

### Remove punctuations

```python
punctuations = '''\n!()-[]{};:'"\, <>./?@#$%^&*_~’'''

for character in text: 
  if character in punctuations: 
    text = text.replace(character, ' ')
```

### To lower case 

```python
text = text.lower() 
```

```python
print(text)
```

### Tokenizations

```python
tokens = text.split(' ')
```

```python
print(tokens)
```

```python
from collections import Counter 
print(Counter(tokens).most_common(50))
```

### Filter Stopwords

```python
# Stopwords 
url = 'https://raw.githubusercontent.com/theleadio/datascience_demo/master/stopwords.txt'
response = requests.get(url)
stopwords = response.text.splitlines()
```

```python
tokens = [t for t in tokens if t not in stopwords and len(t) > 0]
```

```python
print(Counter(tokens).most_common(50))
```
