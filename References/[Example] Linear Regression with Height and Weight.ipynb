```python
import pandas as pd 
import seaborn as sns 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/theleadio/datascience_demo/master/weight-height.csv')
df.head(10)
```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.regplot(data=df, x="Height", y = "Weight")
```

```python
df["Weight"] = df["Weight"].mul(0.453592)
df["Height"] = df["Height"].mul(2.54)

df
```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.regplot(data=df, x="Height", y = "Weight")
```

```python
sns.regplot(data=df.sample(200), x="Height", y = "Weight")
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 
```

```python
# Use case: Predict weight using height 
feature = ['Height']

train_data = df[feature]
train_target = df['Weight']
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_data, train_target, test_size=0.3, random_state=1)
```

```python
model = LinearRegression()
model.fit(X_train.values, y_train)
```

```python
model.score(X_test.values, y_test)
```

```python
model.predict([[170]])
```
