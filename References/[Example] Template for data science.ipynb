# Machine Learning Model

### Classifiers

#### Decision tree model

```python
# Extract a separate dataframe. 
df2         = df.copy()

# Check if there are any NaN elements in each column.
for col in df2:
    print(col, ' : ', df2[col].isna().any())
```

```python
# Drop any NaN values if NaN is observed. 
df2         = df.dropna()

# Get all column headers
df2.columns.values
```

```python
# Columns to be trained. (Include the required features.)
train_column  = ['features']

# Create training data. (Use to develop the tree model.)
train_data    = df2[train_column]

# Degine the training labels. (Condition/Items that to be predicted.)
train_label   = df2['features']

# Split our dataset into training and testing. (Test size = 30% test, 70% training)
x_train, x_test, y_train, y_test  = train_test_split(train_data, train_label, test_size = 0.3, random_state = 1)

# Build decision tree model.
model        = tree.DecisionTreeClassifier(max_depth = 3)

# Train decision tree model.
model.fit(x_train, y_train)
```

```python
# Extract data from decision tree model.
tree_column  = list(train_data.columns)
dot_data     = tree.export_graphviz(model, out_file = None, feature_names = tree_column, class_names = ['No', 'Yes'], filled = True, rounded = True)

# Publish decision tree.
graph        = graphviz.Source(dot_data)
graph
```

```python
# Get predictions using test set.
y_pred       = model.predict(x_test)

# Check the accuracy of decision tree model.
print('Accuracy : ', metrics.accuracy_score(y_test,y_pred))
```
