# Load necessary libraries

# Load data into a dataframe

DataFrame is a table-like data structure with columns different types. You can think of it like an Excel spreadsheet or SQL table. We usually use `df` to represent a dataframe. 

The dataset used https://www.kaggle.com/datasets/paradisejoy/top-hits-spotify-from-20002019

```python
df = pd.read_csv("https://raw.githubusercontent.com/Mund99/DSUB_modules/73cb03c05fda34de4e9afeaa8ffeec4823e20b8e/songs_normalize.csv")
```
