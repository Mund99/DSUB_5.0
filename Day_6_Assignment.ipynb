#Challenge (Second chart) 
- Finding the top 10 most popular songs according to the popularity rank

```python
# Obtain the required data feature 
a = df[['artist','song','popularity']]
```

```python
a.loc[:,['artist']]
```

```python
# Create new column with concatenating artist name and the song 
a['artist, song'] = a[["artist", "song"]].apply(", ".join, axis=1)
a
```

```python
# Obtain the data feature of 'artist, song' and 'popularity' onlly 
a = a.loc[:,['artist, song','popularity']]
a = a.sort_values(by=['popularity'],ascending=False).iloc[0:10]
a
```

```python
# Bar plot 
ay = sns.barplot(x='artist, song',y= 'popularity',data=a)

# Set Labels 
sns.set(rc={'figure.figsize':(15,10)})
plt.xticks(fontsize=10,rotation=45)

ay
```

# Chart 3 
- Selecting the particular artist
- Generate the chart of the popularity of songs generated in each year 

```python
# Variable artist 
artist = "Avicii"

# Filter the dataset with the particular artist
# Extract required data feature and sort the values by year 
b = df[(df['artist'] == artist)]
b = b[['song','year','popularity']].sort_values(by=['year'],ascending=False)
b
```

```python
# Line plot 
ay = sns.lineplot(x='year',y= 'popularity',data=b)

# Set Labels 
sns.set(rc={'figure.figsize':(10,5)})
plt.title("Average popularity for the songs by " + artist + " from 2000 - 2019")

ay
```
